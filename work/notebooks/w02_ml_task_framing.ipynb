## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

## ML Task Type

My lane is a ranking/scoring problem. The goal is to assign each page a score based on how likely it is to need a content refresh. This helps rank pages so content teams can review the most important pages first instead of checking every page manually.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## Target or Proxy

The target is a content refresh score that estimates how likely a page needs to be updated. Since there is no direct label saying "needs refresh," the model will use a proxy based on observed search and content signals, such as impressions, CTR, content age, and other page metrics.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success Metric

The main success metric is Precision@K. A higher Precision@K means more of the top-ranked pages actually need attention. This is important because content teams usually review only a limited number of pages, so the highest-ranked recommendations should be as accurate as possible.

Each row in the dataframe represents one webpage. Each page contains search and content-related metrics such as impressions, CTR, content age, and other features. These page-level signals will be used to estimate which pages should be prioritized for content refresh.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML Beats a Fixed Rule

A fixed rule cannot consider all page signals at the same time. Machine learning can learn patterns from multiple features such as impressions, CTR, content age, and engagement to rank pages more accurately. This provides better decision support than using a single manually defined rule.
